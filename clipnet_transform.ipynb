```python
import numpy as np
from pyfaidx import Fasta
```

```python
data = np.load("output/p16_tss.npz", allow_pickle=True)
importance_scores = data["arr_0"]   # shape: (N, 4, 1000)
print("✅ importance shape:", importance_scores.shape)

# ==== 2. from fasta to one-hot ====
def one_hot_encode(seq):
    mapping = {'A':0, 'C':1, 'G':2, 'T':3}
    onehot = np.zeros((len(seq), 4), dtype=np.float32)
    for i, base in enumerate(seq.upper()):
        if base in mapping:
            onehot[i, mapping[base]] = 1.0
    return onehot

fasta = Fasta("data/p16_sen_sharp_TSS_sequences_clipnet.fasta")
one_hot = np.array([one_hot_encode(str(fasta[name])) for name in fasta.keys()])
print("✅ one-hot raw shape:", one_hot.shape)

# change the dimention into (N, 4, L)
one_hot = np.transpose(one_hot, (0, 2, 1))
print("✅ one-hot final shape:", one_hot.shape)

# ==== 3. save .npz ====
np.savez("importance_scores.npz", importance_scores=importance_scores)
np.savez("one_hot.npz", one_hot=one_hot)
```

```python
# 如果之前生成了 one_hot.npz 和 importance_scores.npz
ohe = np.load("one_hot.npz", allow_pickle=True)
shap = np.load("importance_scores.npz", allow_pickle=True)

np.savez("ohe.npz", arr_0=ohe[list(ohe.keys())[0]])
np.savez("shap.npz", arr_0=shap[list(shap.keys())[0]])
```
